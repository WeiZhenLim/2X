```python
import pandas as pd
import tldextract
```

```python
# update the filename
excel_file = "Company Domain - Third Batch" 

df = pd.read_csv(excel_file + ".csv")
df
```

```python
df["Subdomain"] = df["Company Domain"].apply(lambda x : tldextract.extract(x).subdomain)
df["Domain"] = df["Company Domain"].apply(lambda x : tldextract.extract(x).domain)
df["Suffix"] = df["Company Domain"].apply(lambda x : tldextract.extract(x).suffix)
df
```

```python
# update the filename
new_File = excel_file + "-Output"

writer = pd.ExcelWriter(new_File + ".xlsx")
df.to_excel(writer, index = None)
writer.close()
```
